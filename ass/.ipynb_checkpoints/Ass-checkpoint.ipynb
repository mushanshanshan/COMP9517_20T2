```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
image_I=cv2.imread('Particles.png',0)
print(image_I.shape)
```

```python
plt.imshow(image_I,'gray') 
```

```python
def neighbourhood(i,j,N,image):
    l=N//2
    temp=np.zeros(N*N)
    count=0
    for m in range(-l,l+1):
        for n in range(-l,l+1):
            temp[count]=image[i+m,j+n]
            count+=1
    return temp
```

```python
def max_filtered(N,image):
    image_A=np.zeros_like(image)
    l=N//2
    image_pad=cv2.copyMakeBorder(image,l,l,l,l, cv2.BORDER_REPLICATE)    
    for i in range(l,image.shape[0]+l):
        for j in range(l,image.shape[1]+l):
            temp=neighbourhood(i,j,N,image_pad)
            image_A[i-l,j-l]=np.max(temp)
    return image_A               
```

```python
def min_filtered(N,image):
    image_B=np.zeros_like(image)
    l=N//2
    image_pad=cv2.copyMakeBorder(image,l,l,l,l,cv2.BORDER_REPLICATE)    
    for i in range(l, image.shape[0]+l):
        for j in range(l,image.shape[1]+l):
            temp=neighbourhood(i,j,N,image_pad)
            image_B[i-l,j-l]=np.min(temp)
    return image_B 
```

```python
#N=int(input("Please input the value of N: "))
N=13
image_A=max_filtered(N,image_I)
image_B=min_filtered(N,image_A)
plt.imshow(image_B, 'gray')
```

```python
def q1(image):
    a=0
    b=255
    c=np.min(image)
    d=np.max(image)
    output=(image-c)*((b-a)/(d-c))+a
    output=np.uint8(output)  
    return output
```

```python
def intensity_histogram(img):
    intensity_dict = {}
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            if img[x,y] not in intensity_dict.keys():
                intensity_dict[img[x,y]] = 1
            else:
                intensity_dict[img[x,y]]+= 1
    return intensity_dict
```

```python
intensity_dict = intensity_histogram(image_B)
plt.bar([_ for _ in intensity_dict.keys()],[_ for _ in intensity_dict.values()], color='g') 
plt.show() 
```
